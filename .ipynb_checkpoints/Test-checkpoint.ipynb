```python
tot = nb.frame(nb.totals(50000), 'tot')
```

# coll_newspaper

```python
c1 = coll_newspaper('krigen', before=5, after = 5, datefrom="1945-04-09", dateto="1945-04-20")
c2 = coll_newspaper('krigen', before=10, after = 10, datefrom="1945-04-09", dateto="1945-04-20")
```

```python
c1 = coll_newspaper('krig', before=5, after = 5, datefrom="1930-04-09", dateto="1940-04-20")
c2 = coll_newspaper('krig', before=10, after = 10, datefrom="1930-04-09", dateto="1940-04-20")
```

```python
coll1 = nb.frame(nb.frame(c1).transpose(),"freq doc dist".split())
coll2 = nb.frame(nb.frame(c2).transpose(),"freq doc dist".split())
```

```python
coll = pd.DataFrame()
```

```python
import numpy as np
```

```python
coll['small'] = coll1.freq
coll['large'] = coll2.freq
coll['small_rel'] = coll.small/coll.small.sum()
coll['large_rel'] = coll.large/coll.large.sum()
coll['ratio_rel'] = coll.small_rel/coll.large_rel
```

```python
coll['ratio_rel'] = coll.small_rel**0.08*coll.small_rel/coll.large_rel
```

```python
coll['ratio'] = [(x[1]['small']/x[1]['large']) for x in coll.iterrows()]
```

```python
coll.sort_values(by='ratio_rel', ascending=False)[:50]
```

```python
coll[coll.small < coll.large].sort_values(by='ratio', ascending=False)[:30]
```

```python
coll['nb'] = coll.small/tot.tot
```

```python
coll.sort_values(by='nb', ascending=False)[:30]
```
