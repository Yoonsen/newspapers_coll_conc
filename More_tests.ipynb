```python
dfrom = "2003-01-01"
dto = "2005-12-31"
collword = 'spiste'
d1 = coll_newspaper( collword, before = 5, after = 0, datefrom= dfrom, dateto= dto)
print('en...')
d2 = coll_newspaper( collword, before = 10, after = 0, datefrom= dfrom, dateto= dto)
print('to...')
d3 = coll_newspaper( collword, before = 15, after = 0, datefrom= dfrom, dateto= dto)
```

```python
coll1 = nb.frame(nb.frame(c1).transpose(),"freq doc dist".split())
coll2 = nb.frame(nb.frame(c2).transpose(),"freq doc dist".split())
coll3 = nb.frame(nb.frame(c3).transpose(),"freq doc dist".split())
```

```python
doll1 = nb.frame(nb.frame(d1).transpose(),"freq doc dist".split())
doll2 = nb.frame(nb.frame(d2).transpose(),"freq doc dist".split())
doll3 = nb.frame(nb.frame(d3).transpose(),"freq doc dist".split())
```

```python
coll = pd.DataFrame()
doll = pd.DataFrame()
```

```python
import numpy as np
```

```python
coll['small'] = coll1.freq
coll['large'] = coll2.freq
coll['vlarge'] = coll3.freq
coll['small_rel'] = coll.small/coll.small.sum()
coll['large_rel'] = coll.large/coll.large.sum()
coll['vlarge_rel'] = coll.vlarge/coll.vlarge.sum()
coll['ratio_rel'] = coll.small_rel/coll.large_rel
coll['vratio_rel'] = coll.small_rel/coll.vlarge_rel
```

```python
doll['small'] = doll1.freq
doll['large'] = doll2.freq
doll['vlarge'] = doll3.freq
doll['small_rel'] = doll.small/doll.small.sum()
doll['large_rel'] = doll.large/doll.large.sum()
doll['vlarge_rel'] = doll.vlarge/doll.vlarge.sum()
doll['ratio_rel'] = doll.small_rel/doll.large_rel
doll['vratio_rel'] = doll.small_rel/doll.vlarge_rel
```

```python
coll['drop'] = (coll.small_rel/coll.large_rel)/(coll.large_rel/coll.vlarge_rel)
```

```python
doll['drop'] = (doll.small_rel/doll.large_rel)/(doll.large_rel/doll.vlarge_rel)
```

```python
coll['ratio_rel'] = (coll.small_rel**0.01)*coll.small_rel/coll.large_rel
coll['vratio_rel'] = (coll.small_rel**0.01)*coll.small_rel/coll.vlarge_rel
```

```python
doll['ratio_rel'] = (doll.small_rel**0.01)*doll.small_rel/doll.large_rel
doll['vratio_rel'] = (doll.small_rel**0.01)*doll.small_rel/doll.vlarge_rel
```

```python
coll['ratio'] = coll.small/coll.large
coll['vratio'] = coll.small/coll.vlarge
```

```python
doll['ratio'] = doll.small/coll.large
doll['vratio'] = doll.small/coll.vlarge
```

```python
coll[coll.small > 2].sort_values(by='ratio', ascending=False)[:20]
```

```python
doll[doll.small > 2].sort_values(by='ratio', ascending=False)[:20]
```

```python
coll[coll.small < coll.large].sort_values(by='ratio_rel', ascending=False)[:30]
```

```python
doll[doll.small < doll.large].sort_values(by='ratio_rel', ascending=False)[:30]
```

```python
coll['nb'] = coll.small/tot.tot
```

```python
doll['nb'] = doll.small/tot.tot
```

```python
coll.sort_values(by='nb', ascending=False)[:30]
```

```python
doll.sort_values(by='nb', ascending=False)[:30]
```
