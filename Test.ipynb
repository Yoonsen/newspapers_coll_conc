# coll_newspaper

```python
c1 = coll_newspaper('krigen', before=5, after = 5, datefrom="1945-04-09", dateto="1945-04-20")
c2 = coll_newspaper('krigen', before=10, after = 10, datefrom="1945-04-09", dateto="1945-04-20")
```

```python
dfrom = "2011-07-23"
dto = "2011-07-23"
collword = 'terror'
c1 = coll_newspaper( collword, before=5, after = 5, datefrom= dfrom, dateto= dto)
c2 = coll_newspaper( collword, before=10, after = 10, datefrom= dfrom, dateto= dto)
c3 = coll_newspaper( collword, before=20, after = 20, datefrom= dfrom, dateto= dto)
```

```python
coll1 = nb.frame(nb.frame(c1).transpose(),"freq doc dist".split())
coll2 = nb.frame(nb.frame(c2).transpose(),"freq doc dist".split())
coll3 = nb.frame(nb.frame(c3).transpose(),"freq doc dist".split())
```

```python
coll = pd.DataFrame()
```

```python
import numpy as np
```

```python
coll['small'] = coll1.freq
coll['large'] = coll2.freq
coll['vlarge'] = coll3.freq
coll['small_rel'] = coll.small/coll.small.sum()
coll['large_rel'] = coll.large/coll.large.sum()
coll['vlarge_rel'] = coll.vlarge/coll.vlarge.sum()
coll['ratio_rel'] = coll.small_rel/coll.large_rel
coll['vratio_rel'] = coll.small_rel/coll.vlarge_rel
```

```python
coll['drop'] = (coll.small_rel/coll.large_rel)/(coll.large_rel/coll.vlarge_rel)
```

```python
coll['ratio_rel'] = (coll.small_rel**0.01)*coll.small_rel/coll.large_rel
coll['vratio_rel'] = (coll.small_rel**0.01)*coll.small_rel/coll.vlarge_rel
```

```python
coll['ratio'] = coll.small/coll.large
coll['vratio'] = coll.small/coll.vlarge
```

```python
coll[coll.small > 5].sort_values(by='ratio', ascending=False)[:20]
```

```python
coll[coll.small < coll.large].sort_values(by='drop', ascending=False)[:30]
```

```python
coll['nb'] = coll.small/tot.tot
```

```python
coll.sort_values(by='nb', ascending=False)[:30]
```
